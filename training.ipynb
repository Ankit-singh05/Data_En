```python
import pandas as pd

# Create DataFrame
data = {
    "id": [1, 2, 3],
    "name": ["Ankit", "Ravi", "Neha"],
    "salary": [50000, 60000, 55000]
}

df = pd.DataFrame(data)

# Show data
print(df)

# Add new column
df["bonus"] = df["salary"] * 0.10

# Filter rows
high_salary = df[df["salary"] > 55000]

print("\nEmployees with salary > 55000:")
print(high_salary)
```
